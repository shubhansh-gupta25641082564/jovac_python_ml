```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
print("### Summary statistics\n")
print(df.describe(include='all'), end="\n\n")
```

```python
print("### Missing values by column\n")
print(df.isnull().sum(), end="\n\n")
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time')
plt.title('Tip vs Total Bill, colored by Time of Day')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='day', y='total_bill')
plt.title('Total Bill Distribution by Day of Week')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4, 4))
sns.countplot(data=df, x='smoker')
plt.title('Count: Smoker vs NonSmoker')
plt.tight_layout()
plt.show()
```

---
### 3. Documented Findings
---

- The scatterplot shows a positive correlation between **total bill** and **tip**: larger bills tend to yield higher tips.  
- **Dinner** (vs Lunch) tends to have higher average bills and larger tip amounts.  
- The boxplot reveals **Friday** has the largest spread in total bills, indicating more variability in spending on that day.  
- About **38%** of patrons in this sample are **smokers**.  

No missing data were found, so no imputation was needed.
